```python
import numpy as np
```

```python
arr = np.random.randint(1,100,5)
arr2 = np.random.randint(1,200,5)
print(arr)
print(type(arr2))

arr3=list(arr2)
print(type(arr3))
```

```python
arr4 = np.array(arr3)
print(arr4)
main = np.concatenate([arr,arr4])
print(main)
len(main)
```

```python
print(l)
print(main)
```
